```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Fixer la graine pour la reproductibilité
np.random.seed(42)

def generate_messy_data(n=1000):
    data = {
        'ID_Client': range(1000, 1000 + n),
        'Age': np.random.choice([25, 35, 45, 150, np.nan, 18], size=n),
        'Revenu_Annuel': np.random.normal(50000, 15000, n),
        'Score_Credit': np.random.randint(300, 850, n),
        'Education': np.random.choice(['Lycée', 'Licence', 'Master', None], n),
        'Ville': np.random.choice(['Paris', 'Lyon', 'Marseille', 'Pris', 'Lyon '], n),
        'Taux_Endettement': np.random.choice(['0.3', '30%', '0.15', '15%', np.nan], n),
        'Date_Ouverture': [datetime(2010, 1, 1) + timedelta(days=np.random.randint(0, 4000)) for _ in range(n)],
        'Genre': np.random.choice(['M', 'F', 'Inconnu'], n),
        'Pret_Accorde': np.random.choice(['Oui', 'Non'], n)
    }

    df = pd.DataFrame(data)

    # Injecter des erreurs spécifiques
    df.loc[0:10, 'Score_Credit'] = -999  # Valeurs impossibles
    df.loc[20:30, 'Revenu_Annuel'] = df.loc[20:30, 'Revenu_Annuel'] / 1000 # Erreur d'unité

    return df

df = generate_messy_data()
df.to_csv('credit_risk_tp.csv', index=False)
print("Dataset 'credit_risk_tp.csv' généré avec succès !")
```

```python
# 2. Nettoyage de la colonne 'Taux_Endettement' (Mélange de string et float)
def clean_debt_ratio(val):
    if pd.isna(val): return np.nan
    val = str(val).replace('%', '')
    val = float(val)
    return val / 100 if val > 1 else val

df['Taux_Endettement'] = df['Taux_Endettement'].apply(clean_debt_ratio)
```

# Étape 1: Correction des fautes

```python
# 3. Correction des fautes de frappe (Ville)
df['Ville'] = df['Ville'].str.strip().replace({'Pris': 'Paris'})
```

# Étape 2 : Gestion des Valeurs Manquantes (Imputation)

```python
# Imputation par la médiane pour l'Age (Ratio - robuste aux outliers)
df['Age'] = df['Age'].fillna(df['Age'].median())

# Imputation par le mode pour l'Education (Ordinal)
mode_edu = df['Education'].mode()[0]
df['Education'] = df['Education'].fillna(mode_edu)

# Suppression des lignes si le Genre est 'Inconnu' (Choix stratégique)
df = df[df['Genre'] != 'Inconnu']
```

```python
# Correction des scores de crédit négatifs (Intervalle)
df.loc[df['Score_Credit'] < 0, 'Score_Credit'] = df['Score_Credit'].median()
```

```python
# Plafonnement de l'âge (Ratio)
df.loc[df['Age'] > 100, 'Age'] = df['Age'].median()
```

```python
# Correction du revenu (si < 1000, on considère que c'est en K€ et on multiplie par 1000)
df.loc[df['Revenu_Annuel'] < 1000, 'Revenu_Annuel'] *= 1000
```

```python
df
```

# Étape 4 : Encodage des Variables Catégorielles

```python
# A. Encodage Ordinal (Education : il y a une hiérarchie)
edu_mapping = {'Lycée': 1, 'Licence': 2, 'Master': 3}
df['Education'] = df['Education'].map(edu_mapping)
```

```python
# B. Encodage Nominal / One-Hot (Ville : pas de hiérarchie)
df = pd.get_dummies(df, columns=['Ville', 'Genre'], drop_first=True)
```

```python
# C. Encodage de la Cible (Target)
df['Pret_Accorde'] = df['Pret_Accorde'].map({'Oui': 1, 'Non': 0})
```

# création de variables(Ingenierie de caracteristiques)

```python
df['Date_Ouverture'] = pd.to_datetime(df['Date_Ouverture'])
reference_date = pd.to_datetime('2026-01-01')
df['Ancienneté_Jours'] = (reference_date - df['Date_Ouverture']).dt.days
df.drop('Date_Ouverture', axis=1, inplace=True)
```
